```python
url = "https://api.open-meteo.com/v1/forecast?latitude=39.9523&longitude=-75.1638&current=temperature_2m,apparent_temperature&hourly=temperature_2m,apparent_temperature,wind_gusts_10m,temperature_120m,temperature_180m&daily=temperature_2m_max,temperature_2m_min&temperature_unit=fahrenheit&timezone=America%2FNew_York"
import requests
response = requests.get(url)
response
```

```python
url = "https://api.open-meteo.com/v1/forecast?latitude=51.4712&longitude=-0.4565&current=temperature_2m,apparent_temperature&hourly=temperature_2m,apparent_temperature,wind_gusts_10m,temperature_120m,temperature_180m&daily=temperature_2m_max,temperature_2m_min&temperature_unit=fahrenheit&timezone=America%2FNew_York&forecast_days=3"
import requests
response = requests.get(url)
response
```
